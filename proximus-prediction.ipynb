### Zip the result folder

```python
import zipfile
import os

def zip_folder(folder_path, zip_path):
    
    Compresses a folder into a ZIP archive.

    :param folder_path: Path to the folder to be compressed.
    :param zip_path: Path to the output ZIP archive.
    

    with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                zipf.write(file_path, os.path.relpath(file_path, folder_path))

zip_folder('results-40/', 'pred_40_epochs.zip')
# zip_folder('ada-image-recognition-fiber/dataset/images/aug_train/', 'aug_img.zip')
# zip_folder('ada-image-recognition-fiber/dataset/labels/aug_train/', 'aug_lbl.zip')
```

### Reformat the outputs to match the required output

```python
import os

# Define the dictionary of classes
class_dict = {0: 'two_screws', 1: 'two_foams', 2: 'two_plastic_covers', 3: 'two_tie_wraps', 4: 'two_rubbers'}

# Define the path to the directory containing the label files
label_dir = '/kaggle/working/results-60/exp/labels/'
img_dir = "/kaggle/input/ada-image-recognition-fiber/dataset/images/test/"

# Get a list of all label files in the directory, sorted alphabetically
img_files = sorted(os.listdir(img_dir))

# Define the output file name
output_file = 'predictions__6.csv'

with open(output_file, "w") as f:
    f.write("image_id_and_subtask,prediction\n")

# Loop through each label file
for img_file in img_files:
    
    file_name = os.path.splitext(os.path.basename(img_file))[0]
    label_file = label_dir + file_name + ".txt"
    
    # Define the output lines for this label file
    output_lines = []
    
    if os.path.isfile(label_file):

        # Open the label file
        with open(label_file, 'r') as f:
            lines = f.readlines()

        classes_found = [int(x.split()[0]) for x in lines]

        for class_label in range(5):
            class_name = class_dict[class_label]
            new_class_val = 1 if classes_found.count(class_label) >= 2 else 0
            output_lines.append(f'{file_name}_{class_name},{new_class_val}\n')
    else:
        for class_label in range(5):
            class_name = class_dict[class_label]
            output_lines.append(f'{file_name}_{class_name},0\n')
        
    # Write the output lines to the output file
    with open(output_file, 'a') as f:
        f.writelines(output_lines)
    
    #for i in output_lines:
        #print(i)
    
    #break
```

### Preapre input files fro training and validation predictions

```python
import os
from sklearn.model_selection import train_test_split

initial_train_path = "/kaggle/input/ada-image-recognition-fiber/dataset/images/train/"
aug_train_path = "/kaggle/input/data-fiber-augs/dataset_aug/images/aug_train/"

initial_train = os.listdir(initial_train_path)
initial_train = list(map(lambda x: initial_train_path + x, initial_train))

aug_train = os.listdir(aug_train_path)
aug_train = list(map(lambda x: aug_train_path + x, aug_train))

all_train_imgs = initial_train + aug_train

train_imgs, val_imgs = train_test_split(all_train_imgs, train_size=0.9, random_state=1)
label_func = lambda x: os.path.splitext(x.replace("images", "labels"))[0] + ".txt"
train_labels, val_labels = list(map(label_func, train_imgs)), list(map(label_func, val_imgs))

with open("train_imgs.txt", "w") as f:
    f.write("\n".join(train_imgs))

with open("val_imgs.txt", "w") as f:
    f.write("\n".join(val_imgs))
```

### Compute training accuracy

```python
import os

def get_classes_found(label_file):
    if os.path.isfile(label_file):
        # Open the label file
        with open(label_file, 'r') as f:
            lines = f.readlines()
        return [int(x.split()[0]) for x in lines]
    else:
        return []

for model in ["40", "50", "60"]:
    for dataset in ["train", "val"]:
        if dataset == "train":
            imgs, labels = (train_imgs, train_labels)
        else:
            imgs, labels = (val_imgs, val_labels)
            
        n_found, all_preds = 0, len(imgs) * 5
        for img_file, label_file in zip(imgs, labels):

            file_name = os.path.splitext(os.path.basename(img_file))[0]
            real_classes = get_classes_found(label_file)
            pred_classes = get_classes_found("results-" + dataset + "-" + model + "/exp/labels/" + file_name + ".txt")

            if len(real_classes) == 0 and len(pred_classes) == 0:
                n_found += 5
                continue
            elif (len(real_classes) != 0 and len(pred_classes) == 0) or (len(real_classes) == 0 and len(pred_classes) != 0):
                continue

            for class_label in range(5):
                if real_classes.count(class_label) >= 2 and pred_classes.count(class_label) >= 2:
                    n_found += 1
                elif real_classes.count(class_label) < 2 and pred_classes.count(class_label) < 2:
                    n_found += 1
        
        print("model:", model, "dataset:", dataset)
        print("n_found:", n_found, "n_all:", all_preds, "Acc:", n_found / all_preds)
        print()
```
